```python
import pandas as pd
```

```python
df = pd.read_csv('nba-players.csv', )
df = df.set_index('name')
df = df.drop('Unnamed: 0', axis=1)
df.head(50)
```

```python
df2 = pd.read_excel('NBA Rookies by Year.xlsx')
df2 = df2[['Name','Year Drafted']]
df2 = df2.set_index('Name')
df2
```

```python
df3 = df.join(df2, how='inner')
df3 = df3.sort_values(by=['Year Drafted'], ascending=False)
df3 = df3.reset_index()
df3 = df3.rename(columns={'index': 'name'})
```
